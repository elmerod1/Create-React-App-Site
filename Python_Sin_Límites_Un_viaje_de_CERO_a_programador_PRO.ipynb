# 1.2.1 Bits y Bytes

Los ordenadores solo tienen dos signos b√°sicos, {0, 1}, pero se pueden combinar en secuencias, as√≠ que no estamos limitados a s√≥lo dos posibles significados. Una variable que s√≥lo puede tomar uno de los dos valores binarios recibe el nombre de bit (acr√≥nimo del ingl√©s ‚ú≠‚ú≠binary digit‚úÆ‚úÆ). Es habitual trabajar con secuencias de bits de tama√±o fijo. Una secuencia de $8$ bits recibe el nombre de byte. Con una secuencia de $8$ bits podemos
representar $256$ significados diferentes. El rango $[0, 255]$ de valores naturales comprende $256$ valores, as√≠ que podemos representar cualquiera de ellos con un patr√≥n de $8$ bits.

Centramos el discurso en secuencias de 8 bits, aunque todo lo que exponemos a continuaci√≥n es v√°lido para secuencias de otros tama√±os. El valor de una cadena de bits $b_7b_6b_5b_4b_3b_2b_1b_0$ es, en un sistema posicional convencional, $\sum_{i=0}^{7} b^i2^i$ As√≠, la secuencia de bits $00001011$ codifica el valor

$0 ¬∑ 2^7 + 0 ¬∑ 2^6 + 0 ¬∑ 2^5 + 0 ¬∑ 2^4 + 1 ¬∑ 2^3 + 0 ¬∑ 2^2 + 1 ¬∑ 2^1 + 1 ¬∑ 2^0 = 8 + 2 + 1 = 11$.

El bit de m√°s a la izquierda recibe el nombre de ‚ú≠‚ú≠bit m√°s significativo‚úÆ‚úÆ y el bit de m√°s a la derecha se denomina ‚ú≠‚ú≠bit menos significativo‚úÆ‚úÆ.

# 1.3 Lenguajes de programaci√≥n

Los programas de computadora, llamados software, son conjuntos de instrucciones que le dicen a la computadora qu√© hacer. Dado que las computadoras no comprenden lenguajes humanos, los programas deben escribirse en lenguajes espec√≠ficos para que puedan ser entendidos por la computadora. Aunque existen muchos lenguajes de programaci√≥n dise√±ados para facilitar la tarea de programar, todos los programas deben ser convertidos a un lenguaje que la computadora pueda procesar.

# 1.3.1 Lenguaje m√°quina

El lenguaje nativo de una computadora, que difiere entre diferentes tipos de computadoras, es su lenguaje de m√°quina, un conjunto de instrucciones primitivas incorporadas. Estas instrucciones se presentan en forma de c√≥digo binario, por lo que si deseas darle una instrucci√≥n a una computadora en su lenguaje nativo, debes ingresar la instrucci√≥n en forma de c√≥digo binario. Por ejemplo, para sumar dos n√∫meros, podr√≠as tener que escribir una instrucci√≥n en c√≥digo binario, como esta: $1101101010011010$.

# 1.3.2 Lenguaje ensamblador

Programar en lenguaje de m√°quina es tedioso y los programas resultantes son dif√≠ciles de leer y modificar. Por esta raz√≥n, se cre√≥ el lenguaje ensamblador como una alternativa. El lenguaje ensamblador utiliza palabras descriptivas cortas llamadas mnemot√©cnicas para representar las instrucciones de lenguaje de m√°quina. Por ejemplo, "add" representa la suma de n√∫meros. Esto hace que la programaci√≥n sea m√°s legible y comprensible.

add <font color="blue">2</font>, <font color="blue">3</font>, result

Los lenguajes **ensambladores** se desarrollaron para facilitar la programaci√≥n. Sin embargo, dado que la computadora no puede entender el lenguaje ensamblador, se utiliza otro programa, llamado ensamblador, para traducir los programas escritos en lenguaje ensamblador al c√≥digo de m√°quina, como se muestra en la figura.

<p style="text-align:center;"><img src="https://imgtr.ee/images/2023/09/07/9c554b37bd9c5258aa22b342d74bfe48.png"  width = "600" height = "150" border="0"/ alt="Logo"></p>

Escribir c√≥digo en lenguaje ensamblador es m√°s f√°cil que en lenguaje de m√°quina, pero a√∫n puede ser tedioso. Cada instrucci√≥n en lenguaje ensamblador corresponde esencialmente a una instrucci√≥n en lenguaje de m√°quina, y se requiere un conocimiento detallado del funcionamiento de la CPU. El lenguaje ensamblador se considera de **bajo nivel** y es altamente dependiente de la arquitectura de la m√°quina en la que se ejecuta.

# 1.3.3 Lenguaje de alto nivel

En la d√©cada de 1950, surgi√≥ una nueva generaci√≥n de lenguajes de programaci√≥n conocidos como lenguajes de alto nivel. Estos lenguajes son independientes de la plataforma, lo que significa que puedes escribir un programa en un lenguaje de alto nivel y ejecutarlo en diferentes tipos de m√°quinas. Los lenguajes de alto nivel son similares al ingl√©s y f√°ciles de aprender y usar. Las instrucciones en un lenguaje de programaci√≥n de alto nivel se llaman declaraciones. Aqu√≠ tienes, por ejemplo, una declaraci√≥n en lenguaje de alto nivel que calcula el √°rea de un c√≠rculo con un radio de 5:

$$area = 5 * 5 * 3.1415$$

Existen muchos lenguajes de programaci√≥n de alto nivel, y cada uno fue dise√±ado para un prop√≥sito espec√≠fico; tal como se muestran a continuaci√≥n:

<p style="text-align:center;"><img src="https://imgtr.ee/images/2023/09/07/e87fab5c534f43c95de6bd3ffdd23dfc.jpeg"  width = "600" height = "" border="0"/ alt="Logo"></p>

Un programa escrito en un lenguaje de alto nivel se llama programa fuente o c√≥digo fuente. Debido a que una computadora no puede entender un programa fuente, este debe ser traducido al c√≥digo de m√°quina para su ejecuci√≥n. La traducci√≥n se puede realizar utilizando otra herramienta de programaci√≥n llamada int√©rprete o compilador.

- Un int√©rprete lee una declaraci√≥n del c√≥digo fuente, la traduce al c√≥digo de m√°quina o c√≥digo de m√°quina virtual y la ejecuta de inmediato (l√≠nea por linea).

- Un compilador traduce todo el c√≥digo fuente en un archivo de c√≥digo de m√°quina, y luego ese archivo de c√≥digo de m√°quina se ejecuta.

# **1.9 Errores de programaci√≥n**

Los errores de programaci√≥n se pueden clasificar en tres tipos: errores de sintaxis, errores de ejecuci√≥n y errores l√≥gicos.

<br>

### 1.9.1 Errores de sintaxis

El error m√°s com√∫n en Python son los **errores de sintaxis**, que se producen cuando el c√≥digo no sigue las reglas de sintaxis del lenguaje. Estos errores pueden ocurrir, por ejemplo, si falta una comilla, si hay una palabra mal escrita o si se omite una puntuaci√≥n necesaria. Python informar√° sobre estos errores y proporcionar√° informaci√≥n sobre d√≥nde y por qu√© ocurrieron. Estos errores son relativamente f√°ciles de detectar y corregir, ya que Python ofrece pistas sobre su ubicaci√≥n y causa.

Por ejemplo, el siguiente c√≥digo tiene un error de sintaxis:

````python
print("Hola Mundo!)
````

**Nota:** La cadena Hola Mundo!, debe cerrarse con comillas de cierre.

### 1.9.2 Errores de tiempo de ejecuci√≥n

Los **errores de tiempo de ejecuci√≥n** son fallos que causan la terminaci√≥n inesperada de un programa durante su ejecuci√≥n. Estos errores surgen cuando el int√©rprete de Python encuentra operaciones imposibles de realizar, generalmente debido a entradas inadecuadas, como ingresar un tipo de dato incorrecto o intentar dividir por cero. Estos errores interrumpen el flujo normal del programa.

Por ejemplo, la expresi√≥n 1/0 en la siguiente declaraci√≥n provocar√≠a un error de tiempo de ejecuci√≥n.

````python
print(1/0)
````

## 1.8.3 Errores l√≥gicos

Los errores l√≥gicos ocurren cuando un programa no funciona como estaba previsto. Errores de esto tipo ocurren por muchas razones diferentes. Por ejemplo, supongamos que escribiste el siguiente programa para convertir una temperatura (35 grados) de Fahrenheit a Celsius.

```
# Convert Fahrenheit to Celsius
print("Fahrenheit 35 is Celsius degree ")
print(5 / 9 * 35 - 32)
```
Obtendr√° Celsius ‚Äì12,55 grados, lo cual est√° mal. Deber√≠a ser 1,66. Para obtener el resultado correcto, debe utilizar 5/9 * (35 ‚Äì 32) en lugar de 5/9 * 35 ‚Äì 32 en la expresi√≥n. Es decir, debe agregar par√©ntesis alrededor de (35 ‚Äì 32) para que Python calcule esa expresi√≥n primero antes de realizar la divisi√≥n. En Python, los errores de sintaxis en realidad se tratan como errores de tiempo de ejecuci√≥n porque el int√©rprete los detecta cuando se ejecuta el programa.

En general, los errores de sintaxis y de tiempo de ejecuci√≥n son f√°ciles de encontrar y corregir, porque Python da indicaciones sobre de d√≥nde provienen los errores y por qu√© son incorrectos. Por otro lado, encontrar errores l√≥gicos puede resultar muy complicado.

Las variables se utilizan para hacer referencia a valores que pueden cambiarse en el programa.

# ***2.6 Asignaciones simult√°neas***

Python tambi√©n admite la asignaci√≥n simult√°nea con una sintaxis como esta:

`var1, var2, ..., varn = exp1, exp2, ..., expn`

Le dice a Python que eval√∫e todas las expresiones de la derecha y las asigne a la variable correspondiente de la izquierda simult√°neamente. Intercambiar valores de variables es una operaci√≥n com√∫n en programaci√≥n y la asignaci√≥n simult√°nea es muy √∫til para realizar esta operaci√≥n. Considere dos variables: $x$ e $y$. ¬øC√≥mo se escribe el c√≥digo para intercambiar sus valores? Un enfoque com√∫n es introducir una variable temporal de la siguiente manera:

````python
x = 1
y = 2
temp = x      # Guardar x en una variable temporal
x = y         # Asigna el valor en y a x
y = temp      # Asigne el valor en temp a y
````

**NOTA:** ¬øQu√© valor tendr√° $y$?

Pero puedes simplificar la tarea usando la siguiente declaraci√≥n para intercambiar los valores de $x$ e $y$.

````python
x, y = y, x     # Intercambiar x con y.
````

# **2.8 Tipos de datos num√©ricos y operadores**

Python tiene dos tipos num√©ricos (enteros y n√∫meros de punto flotante) para trabajar con los operadores <font color="blue"> $+, -, *, /, //, ** y %.$ </font>

La informaci√≥n almacenada en una computadora generalmente se denomina datos. Hay dos tipos de datos num√©ricos: n√∫meros enteros y n√∫meros reales. Los tipos de n√∫meros enteros (int para abreviar) sirven para representar n√∫meros enteros. Los tipos reales sirven para representar n√∫meros con parte fraccionaria. Dentro de la computadora, estos dos tipos de datos se almacenan de manera diferente. Los n√∫meros reales se representan como valores de punto flotante (o flotantes). ¬øC√≥mo le decimos a Python si un n√∫mero es un n√∫mero entero o un flotante?
